### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Calculate average mpg by year and origin
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Step 2: Plot using Seaborn lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Average MPG over Years by Origin')
plt.xlabel('model_year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()

```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python

# Count how many unique model years each car (index name) appears in
car_year_counts = df.groupby(level=0)['model_year'].nunique()

# Filter to only cars that appear in more than one model year
multi_year_cars = car_year_counts[car_year_counts > 1].index

# Filter df to those multi-year cars
filtered_df = df.loc[df.index.isin(multi_year_cars)]

# Group by index level 0 (car name), calculate std, mean, count
stats = (
    filtered_df
    .groupby(level=0)
    .agg(
        mpg_std=('mpg', 'std'),
        mpg_mean=('mpg', 'mean'),
        appearances=('mpg', 'count')
    )
)

# Only keep cars with low variation in mpg
consistent_cars = stats[stats['mpg_std'] < 1.0]

# Find the minimum std
min_std = consistent_cars['mpg_std'].min()

# Select those with the lowest std
lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

# Final sort: most appearances first, then highest average mpg
final_result = lowest_variation_cars.sort_values(
    by=['appearances', 'mpg_mean'],
    ascending=[False, False]
)

# Show desired columns
print(final_result[['appearances', 'mpg_mean']])

```
